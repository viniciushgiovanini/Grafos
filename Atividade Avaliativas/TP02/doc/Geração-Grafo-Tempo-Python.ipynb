```python
# Importando bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# Arrays
completo = [14, 33, 3056, 281205]
direcionado = [0.0, 10, 862, 86412]
circular = [0.0, 10, 666, 65731]

completo_v = [100, 1000, 10000, 100000]
direcionado_v = [100, 1000, 10000, 100000]
circular_v = [100, 1000, 10000, 100000]

# K-conexo:    Tamanho:    100    Tempo:    14ms
# K-conexo:    Tamanho:    1000    Tempo:    33ms
# K-conexo:    Tamanho:    10000    Tempo:    3056ms
# K-conexo:    Tamanho:    100000    Tempo:    281205ms

# Direcionado:    Tamanho:    100    Tempo:    0ms
# Direcionado:    Tamanho:    1000    Tempo:    10ms
# Direcionado:    Tamanho:    10000    Tempo:    862ms
# Direcionado:    Tamanho:    100000    Tempo:    86412ms

# Circular:    Tamanho:    100000    Tempo:    65731ms
# Circular:    Tamanho:    10000    Tempo:    666ms
# Circular:    Tamanho:    1000    Tempo:    10ms
# Circular:    Tamanho:    100    Tempo:    0ms
```

```python
%matplotlib inline
plt.figure(figsize=(10, 7),dpi=100)
plt.plot(circular_v, circular, label='Grafo Circular')
plt.plot(direcionado_v, direcionado, label='Grafo direcionado')
plt.plot(completo_v, completo, label='Grafo de K5')

plt.xlabel('Quantidade de vertices')
plt.ylabel('Tempo de execução em ms')
plt.title('Comparação de tempo por quantidade de vertices')

plt.legend()

plt.show()
```

```python
%matplotlib inline
plt.figure(figsize=(10, 7))
plt.plot(direcionado_v, direcionado, label='Grafo Simples')

plt.xlabel('Quantidade de vertices')
plt.ylabel('Tempo de execução em ms')
plt.title('Comparação de tempo por quantidade de vertices - direcionado')

plt.legend()

plt.show()
```

```python
%matplotlib inline
plt.figure(figsize=(10, 7))
plt.plot(completo_v, completo, label='Grafo de K5')

plt.xlabel('Quantidade de vertices')
plt.ylabel('Tempo de execução em ms')
plt.title('Comparação de tempo por quantidade de vertices - Grafo de K5')

plt.legend()

plt.show()
```
